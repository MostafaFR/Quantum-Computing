        | 1 | 1 | 1 | 
        | 1 | 0 | 1 | 
        
    - L'op√©ration logique XOR est repr√©sent√© par $\oplus$, retourne 1 si les deux op√©randes sont differents, sinon retourne 1.
        <br>Porte XOR : 
        | X | Y | X $\oplus$ Y | 
        |---|---|-------| 
        | 0 | 0 | 0 |

```python
def gate_h(circuit, qubit0, measure=False):
    # Cr√©ation d‚Äôun circuit quantique instance de QuantumCircuit
    if qubit0:
        # Ajouter une porte X pour l‚Äôentr√©e (qubit) 0
        circuit.x(0)
    # Ajouter une porte H pour l‚Äôentr√©e (qubit) 0 => superposition
    circuit.h(0)
    # Mesure du qubit
    if measure:
        circuit.measure(0, 0)
    return circuit


def gate_x(circuit, qubit0, measure=False):
    # Cr√©ation d‚Äôun circuit quantique instance de QuantumCircuit
    if qubit0:
        # Ajouter une porte X pour l‚Äôentr√©e (qubit) 0
        circuit.x(0)
    # Mesure du qubit
    if measure:
        circuit.measure(0, 0)
    return circuit


def gate_cnot(circuit, qubit0, qubit1, measure=False, inverse=False):
    # Cr√©ation d‚Äôun circuit quantique instance de QuantumCircuit
    if qubit0:
        # Ajouter une porte X pour l‚Äôentr√©e (qubit) 0
        circuit.x(0)
    if qubit1:
        # Ajouter une porte X pour l‚Äôentr√©e (qubit) 1
        circuit.x(1)
    # Ajouter une porte CNOT avec 0 comme qubit de control et le qubit 1 comme cible
    if inverse:
        circuit.cx(1, 0)
    else:
        circuit.cx(0, 1)
    # Mesure du qubit
    if measure:
        circuit.measure(0, 0)
        circuit.measure(1, 1)
    return circuit


def gate_swap(circuit, qubit0, qubit1, measure=False, inverse=False):
    # Cr√©ation d‚Äôun circuit quantique instance de QuantumCircuit
    if qubit0:
        # Ajouter une porte X pour l‚Äôentr√©e (qubit) 0
        circuit.x(0)
    if qubit1:
        # Ajouter une porte X pour l‚Äôentr√©e (qubit) 1
        circuit.x(1)
    # Ajouter une porte SWAP avec 0 comme qubit de control et le qubit 1 comme cible
    if inverse:
        circuit.swap(1, 0)
    else:
        circuit.swap(0, 1)
    # Mesure du qubit
    if measure:
        circuit.measure(0, 0)
        circuit.measure(1, 1)
    return circuit


def gate_cswap(circuit, qubit0, qubit1, qubit2, measure=False):
    # Cr√©ation d‚Äôun circuit quantique instance de QuantumCircuit
    if qubit0:
        # Ajouter une porte X pour l‚Äôentr√©e (qubit) 0
        circuit.x(0)
    if qubit1:
        # Ajouter une porte X pour l‚Äôentr√©e (qubit) 1
        circuit.x(1)
    if qubit2:
        # Ajouter une porte X pour l‚Äôentr√©e (qubit) 2
        circuit.x(2)
    # Ajouter une porte CSWAP avec 0 comme qubit de control et les qubits 1 et 2 comme cibles
    circuit.cswap(0, 1, 2)
    # Mesure du qubit
    if measure:
        circuit.measure(0, 0)
        circuit.measure(1, 1)
        circuit.measure(2, 2)
    return circuit


def gate_ccnot(circuit, qubit0, qubit1, qubit2, measure=False):
    # Cr√©ation d‚Äôun circuit quantique instance de QuantumCircuit
    if qubit0:
        # Ajouter une porte X pour l‚Äôentr√©e (qubit) 0
        circuit.x(0)
    if qubit1:
        # Ajouter une porte X pour l‚Äôentr√©e (qubit) 1
        circuit.x(1)
    if qubit2:
        # Ajouter une porte X pour l‚Äôentr√©e (qubit) 2
        circuit.x(2)
    # Ajouter une porte CCNOT avec 0 et 1 comme qubits de control et le qubit 2 comme cible
    circuit.ccx(0, 1, 2)
    # Mesure du qubit
    if measure:
        circuit.measure(0, 0)
        circuit.measure(1, 1)
        circuit.measure(2, 2)
    return circuit


def gate_toffoli(circuit, qubit0, qubit1, qubit2, measure=False):
    # Cr√©ation d‚Äôun circuit quantique instance de QuantumCircuit
    if qubit0:
        # Ajouter une porte X pour l‚Äôentr√©e (qubit) 0
        circuit.x(0)
    if qubit1:
        # Ajouter une porte X pour l‚Äôentr√©e (qubit) 1
        circuit.x(1)
    if qubit2:
        # Ajouter une porte X pour l‚Äôentr√©e (qubit) 2
        circuit.x(2)
    # Ajouter une porte Toffoli avec 0 et 1 comme qubits de control et le qubit 2 comme cible
    circuit.ccx(0, 1, 2)
    # Mesure du qubit
    if measure:
        circuit.measure(0, 0)
        circuit.measure(1, 1)
        circuit.measure(2, 2)
    return circuit
```

```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_histogram, plot_state_qsphere

def get_counts(qc):
    # Simulate the circuit
    backend = Aer.get_backend('qasm_simulator')
    job = execute(qc, backend, shots=2**14)
    result = job.result()
    counts = result.get_counts(qc)
    return counts

def print_counts(counts):
    for key, value in counts.items():
        print(f'{key}: {value}')
```

Un √©tat de Bell peuvent etre cr√©er avec une porte H et un CNOT. La matrice de la porte H √©tant 
$$
\begin{pmatrix}
1 & 1 \\
1 & -1
\end{pmatrix}
$$
Et ces √©tats de Bell √©tant non deterministe peuvent donc etre pr√©definies, il suffit de connaitre le premier qubit ou le second qubit et on peut determiner l'√©tat correspondant. En l'occurcence voici les √©tat de Bell et leurs correspondance :
|$\ket{00}$|$\ket{01}$|$\ket{10}$|$\ket{11}$|
|---|---|---|---|
| $\frac{1}{\sqrt{2}}(\ket{00} + \ket{11})$ | $\frac{1}{\sqrt{2}}(\ket{01} + \ket{10})$ | $\frac{1}{\sqrt{2}}(\ket{00} - \ket{11})$ | $\frac{1}{\sqrt{2}}(\ket{01} - \ket{10})$ |

2. Mesurer sur un simulateur local un √©tat de Bell et afficher les diff√©rents visuels

```python
#Cr√©ation d‚Äôun circuit quantique instance de QuantumCircuit
circ = QuantumCircuit(2)
#Ajouter une porte H pour l‚Äôentr√©e (qubit) 0 => superposition
circ.h(0)
# Ajouter l‚Äôop√©ration CNOT avec 0 comme qubit de control et le qubit 1 comme cible
# => √©tat de Bell
circ.cx(0, 1)
# Mesure du qubit
circ.measure_all()
# Afficher le circuit
circ.draw('mpl')
```

```python
Aer.backends()
simulator = Aer.get_backend('aer_simulator')
compiled_circuit = transpile(circ, simulator)

job = simulator.run(compiled_circuit, shots=1000)
result = job.result()
print(get_counts(circ))
plot_histogram(get_counts(circ), color='midnightblue', title="Simulateur : Mesure des √©tat de Bell")
```
